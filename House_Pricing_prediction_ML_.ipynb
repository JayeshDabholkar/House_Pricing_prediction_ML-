```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("housing.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data.dropna(inplace= True)
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split
X = data.drop(['median_house_value'], axis = 1)
y = data['median_house_value']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
train_data = X_train.join(y_train)
train_data
```

```python
train_data.hist()
```

```python
train_data.corr()
```

```python
plt.figure(figsize= (12,6))
sns.heatmap(train_data.corr(), annot = True, cmap = 'YlGnBu')
```

```python
#using log distribution data can be normalized and + 1 for avoid zero value

train_data['total_rooms'] = np.log(train_data['total_rooms']+1)
train_data['total_bedrooms'] = np.log(train_data['total_bedrooms']+1)
train_data['population'] = np.log(train_data['population']+1)
train_data['households'] = np.log(train_data['households']+1)
```

```python
train_data.hist()
```

```python
#categorical variable
train_data['ocean_proximity'].value_counts()
```

```python
#changing it to binary

train_data = train_data.join(pd.get_dummies(train_data['ocean_proximity'])).drop(['ocean_proximity'], axis = 1)
```

```python
train_data
```

```python
plt.figure(figsize = (15,7))
sns.heatmap(train_data.corr(), annot = True, cmap = 'YlGnBu')
```

```python
#visualise the prices of costal and inland property

plt.figure(figsize=(15,9))
sns.scatterplot(x = train_data['latitude'], y = train_data['longitude'], hue = train_data['median_house_value'], palette = 'coolwarm')
```

```python
#feature engineering

train_data['bedrooms_ratio'] = train_data['total_bedrooms'] / train_data['total_rooms']
train_data['household_rooms'] = train_data['total_rooms'] / train_data['households']
```

```python
plt.figure(figsize = (15,7))
sns.heatmap(train_data.corr(), annot = True, cmap = 'YlGnBu')
```

```python
from sklearn.linear_model import LinearRegression
X_train, y_train = train_data.drop(['median_house_value'], axis = 1) , train_data['median_house_value']
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
test_data = X_test.join(y_test)

test_data['total_rooms'] = np.log(test_data['total_rooms']+1)
test_data['total_bedrooms'] = np.log(test_data['total_bedrooms']+1)
test_data['population'] = np.log(test_data['population']+1)
test_data['households'] = np.log(test_data['households']+1)

test_data = test_data.join(pd.get_dummies(test_data['ocean_proximity'])).drop(['ocean_proximity'], axis = 1)

test_data['bedrooms_ratio'] = test_data['total_bedrooms'] / test_data['total_rooms']
test_data['household_rooms'] = test_data['total_rooms'] / test_data['households']
```

```python
X_test, y_test = test_data.drop(['median_house_value'], axis = 1) , test_data['median_house_value']
```

```python
reg.score(X_test, y_test)
```

```python
#random forest algo

from sklearn.ensemble import RandomForestRegressor
random = RandomForestRegressor()
random.fit(X_test, y_test)
```

```python
random.score(X_test, y_test)
```

```python
#evaluating the model

from sklearn.model_selection import GridSearchCV

random = RandomForestRegressor()

param_grid = {
    "n_estimators" : [100,200,300],
    "min_samples_split" : [2, 4, 6],
    "max_depth" : [None, 3, 6]
}

grid_search = GridSearchCV(random, param_grid, cv= 5, scoring = 'neg_mean_squared_error', return_train_score = True)

grid_search.fit(X_train, y_train)
```

```python
grid_search.best_estimator_
```

```python
grid_search.best_estimator_.score(X_test, y_test)
```
